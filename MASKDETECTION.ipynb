```python
import numpy as np
import cv2
import os
```

```python
images = []
targets = []
```

```python
with_mask_path  = 'data/with_mask/'
without_mask_path = 'data/without_mask/'
```

```python
content = os.listdir(with_mask_path)
error_count = 1
for image in content:
    try:
        filename = with_mask_path + image
        #print(filename)
        img = cv2.imread(filename)
        #print("Img : ",img)
        img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        resized_img = cv2.resize(img_grey, (100, 100))
        images.append(resized_img)
        targets.append(0)
    except:
        print("Filename Error", error_count)
        error_count += 1
print("Completed")
```

```python
content = os.listdir(without_mask_path)
error_count = 1
for image in content:
    try:
        filename = without_mask_path + image
        #print(filename)
        img = cv2.imread(filename)
        #print("Img : ",img)
        img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        resized_img = cv2.resize(img_grey, (100, 100))
        images.append(resized_img)
        targets.append(1)
    except:
        print("Filename Error", error_count)
        error_count += 1
print("Completed")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(images, targets)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Activation, Dropout
from keras.optimizers import SGD
from keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()
```

```python
model.add(Conv2D(200, (3,3), input_shape=(100, 100, 1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(100, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Dropout(0.3))
model.add(Flatten())

model.add(Dense(50, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
cp = ModelCheckpoint('model-best', verbose=0, save_best_only=True)
model.fit(X_train, Y_train, epochs = 5, callbacks=[cp], validation_split=0.2)
```

```python
face_detect = cv2.CascadeClassifier('C:/Users/K MANGOTRA/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml')
```

```python
source = cv2.VideoCapture(0)
while 1:
    not_to_use, image = source.read()
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_detect.detectMultiScale(gray, 1.5, 1)
    for (x, y, w, h) in faces:
        face_roi = gray[y:y+w, x:x+h]
        resized_image = cv2.resize(face_roi, (100, 100))
        normalized_image = resized_image/255
        reshaped_face = np.reshape(normalized_image, (1, 100, 100, 1))
        result = mod.predict(reshaped_face)[0]
        #print("Result : ", result)
        if result[0] >  result[1]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(image, "Safe", (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        if result[1] >  result[0]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)
            cv2.putText(image, "Not Safe", (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        
            
    cv2.imshow('Verzeo Project', image)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
source.release()
```
